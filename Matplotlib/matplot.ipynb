```python
import matplotlib.pyplot as plt
x = [33,44,55,78,55,66]
y = ["part1","part2","part3","part4","part5","part6"]
colors = ["red","green","blue","yellow","orange","pink"]
plt.bar(y,x, color = colors,edgecolor = "black")
plt.xlabel("parts of a movie", fontsize = 12)
plt.ylabel("popularity",fontsize = 12)
plt.title ("movie chart", fontsize = 15)
plt.show()
```

```python
x= ["day1","day2","day3","day4"]
y = [200,300,400,100]
y1 = [188,380,450,22]
plt.plot (x,y, marker="D",ls = "--",color = "black",label = "week1")
plt.plot (x,y1, marker="D",ls = "-",color = "blue", label = "week2")
plt.legend()
plt.show()
```
