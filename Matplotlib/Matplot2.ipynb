```python
import matplotlib.pyplot as plt

brands = ["oneplus","apple","samsung","nokia","redmi"]
x= [22,35,30,3,10]
c = ["red","silver","blue","red","green"]
ex= [0,0,0,0,0.2]
plt.pie(x,labels = brands, colors = c, explode = ex,autopct = "%f")
plt.show()
```
